```python
# Demonstrates how to value a portfolio and how to convert in different currencies
from Portfolios import Portfolio, PortfolioValuator
from TimeSeries import TimeSeries
from pmdata import fx_convert
from datetime import date

p:Portfolio = Portfolio()
p.buy('MSFT', 10)
p.buy('CSCO', 20)
v:PortfolioValuator = p.valuator()
valdate = v.get_end_date()
val_usd = v.get_valuation(valdate, "adjusted_close", "USD")
print("Portfolio valuation as of {date}: {value:.2f} {currency}".format(date=valdate, value=val_usd, currency="USD"))
# To get the valuation into a different currency, just pass the currency in the valuation call
val_chf = v.get_valuation(valdate, "adjusted_close", "CHF")
print("Portfolio valuation as of {date}: {value:.2f} {currency}".format(date=valdate, value=val_chf, currency="CHF"))
# or just convert the USD value
print("Portfolio valuation as of {date}: {value:.2f} {currency}".format(date=valdate, value=fx_convert(val_usd, 'USD', 'CHF', date(2022,11,28)), currency="CHF"))
```
