```python
# Sample time series
# This sample code loads 30+ years of data from Goodyear and displays its adjusted close informatoin
# Notice how weekend data is forward filled.
# You can play with the dates to zoom in and out in the data

from pmdata import get_timeseries, get_ticker
import matplotlib.pyplot as plt
import pandas as pd

# This is how we are getting the time series of a stock
ts = get_timeseries('GT.US', "adjusted_close")
ticker = get_ticker('GT.US')
# We are getting access to the underlying Numpy array in order to graph it
vector = ts.get_full_time_series()
# We are getting access to the dates in order to display them on the horizontal axis
dates = pd.date_range(ts.get_start_date(), ts.get_end_date(), freq='D')
# And we graph the data with style
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.set_title(ticker.name)
ax.set_ylabel(ticker.currency.get_identifier())
ax.plot(dates, vector, linewidth=1)
plt.show()
```

```python
from pmdata import get_timeseries, get_ticker
from datetime import date
import matplotlib.pyplot as plt
import pandas as pd

# This is how we are getting the time series of a stock
ts = get_timeseries('C40.PA', "adjusted_close")
# We are getting access to the underlying Numpy array in order to graph it
vector = ts.cut2dates(date(2022,1,1), ts.get_end_date())
# We are getting access to the dates in order to display them on the horizontal axis
dates = pd.date_range(date(2022,1,1), ts.get_end_date(), freq='D')
# And we graph the data with style
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.set_title("Amundi CAC 40 ESG")
ax.set_ylabel("EUR")
ax.set_xlabel("Year 2022")
ax.set_xticks([])
ax.plot(dates, vector.get_full_time_series(), linewidth=1)
```
